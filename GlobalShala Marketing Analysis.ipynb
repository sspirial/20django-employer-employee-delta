# Task
Analyze the Facebook ad campaign data for the Superhero U event from the file "/content/drive/MyDrive/Marketing Team Data - Sheet1 (1).csv" to identify at least two campaigns to recommend discontinuing in order to cut costs.

## Load the data

### Subtask:
Load the data from the provided CSV file into a pandas DataFrame.


**Reasoning**:
Load the data from the CSV file into a pandas DataFrame and display the head and data types.



```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Marketing Team Data - Sheet1 (1).csv')
display(df.head())
display(df.info())
```

## Analyze campaign performance

### Subtask:
Calculate relevant metrics like cost per conversion, return on ad spend (ROAS), or any other metric that helps identify underperforming campaigns.


**Reasoning**:
Convert the relevant cost columns to numeric, calculate CPULC, Unique CTR, and conversion rate, and then display the updated DataFrame with the new metrics.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Convert cost columns to numeric
for col in ['Amount Spent in INR', 'Cost Per Click (CPC)', 'Cost per Result (CPR)']:
    df[col] = df[col].astype(str).str.replace(r'[$,]', '', regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Calculate Cost Per Unique Link Click (CPULC)
df['Cost Per Unique Link Click (CPULC)'] = df['Amount Spent in INR'] / df['Unique Link Clicks (ULC)']

# Calculate Unique Click-Through Rate (Unique CTR)
# Add a small epsilon to the denominator to avoid division by zero
df['Unique Click-Through Rate (Unique CTR)'] = (df['Unique Clicks'] / (df['Reach'] + np.finfo(float).eps)) * 100

# Calculate the conversion rate
# Add a small epsilon to the denominator to avoid division by zero
df['Conversion Rate'] = (df['Unique Link Clicks (ULC)'] / (df['Unique Clicks'] + np.finfo(float).eps)) * 100

# Display the DataFrame with the new calculated metrics
display(df.head())

# Visualize CPULC by Campaign Name
plt.figure(figsize=(12, 6))
sns.barplot(x='Campaign Name', y='Cost Per Unique Link Click (CPULC)', data=df)
plt.xticks(rotation=90)
plt.title('Cost Per Unique Link Click (CPULC) by Campaign Name')
plt.tight_layout()
plt.show()

# Visualize Conversion Rate by Campaign Name
plt.figure(figsize=(12, 6))
sns.barplot(x='Campaign Name', y='Conversion Rate', data=df)
plt.xticks(rotation=90)
plt.title('Conversion Rate by Campaign Name')
plt.tight_layout()
plt.show()
```

## Identify campaigns for discontinuation

### Subtask:
Based on the analysis, identify at least two campaigns that are performing poorly according to the chosen metrics.


**Reasoning**:
Sort the DataFrame by 'Cost Per Unique Link Click (CPULC)' and 'Conversion Rate' to identify poorly performing campaigns based on high cost per click and low conversion rate, then identify the campaigns that appear in the poor performance lists and store them in a list.



```python
df_sorted_cpulc = df.sort_values(by='Cost Per Unique Link Click (CPULC)', ascending=False)
display(df_sorted_cpulc[['Campaign Name', 'Cost Per Unique Link Click (CPULC)']].head())

df_sorted_conversion = df.sort_values(by='Conversion Rate', ascending=True)
display(df_sorted_conversion[['Campaign Name', 'Conversion Rate']].head())

# Based on the top of the sorted lists, identify campaigns with high CPULC and low Conversion Rate
poor_performing_campaigns = ['SHU_Students(Australia)', 'SHU_Students(UAE)']
```

## Summarize findings

### Subtask:
Present the findings and recommend the campaigns for discontinuation with justification based on the analysis.


**Reasoning**:
Print the list of recommended campaigns for discontinuation, then filter and display the relevant rows from the original dataframe for these campaigns, and finally provide justification for each discontinuation based on the calculated metrics.



```python
print("Recommended campaigns for discontinuation based on poor performance:")
for campaign in poor_performing_campaigns:
    print(f"- {campaign}")

# Access and display rows for the recommended campaigns
recommended_campaigns_df = df[df['Campaign Name'].isin(poor_performing_campaigns)]
display(recommended_campaigns_df)

# Provide justification for discontinuation
print("\nJustification for Discontinuation:")
for index, row in recommended_campaigns_df.iterrows():
    campaign_name = row['Campaign Name']
    cpulc = row['Cost Per Unique Link Click (CPULC)']
    conversion_rate = row['Conversion Rate']
    print(f"\nCampaign: {campaign_name}")
    print(f"  - Justification: This campaign is recommended for discontinuation due to a high Cost Per Unique Link Click (CPULC) of {cpulc:.2f} and a low Conversion Rate of {conversion_rate:.2f}%. These metrics indicate that the campaign is inefficient in driving conversions at a reasonable cost.")
```

*   "SHU\_Students(UAE)" had CPULC values between \$5.23 and \$11.71, with Conversion Rates ranging from 40.20% to 62.50%.
